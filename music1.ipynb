__Break the .mp3 file into chunks__

```python
from pydub import AudioSegment
from pydub.utils import make_chunks
```

```python
seconds = 3 # segment length in seconds
file_count= 16 # total number of files (training + validation)
segment_len = 1000 *seconds # 5 second segments, calculated in milliseconds
```

```python
audio_track = AudioSegment.from_mp3("flute.mp3")
segments = make_chunks(audio_track, segment_len)
```

```python
len(segments)
```

```python

# Export all segments as .wave files

for i, segment in enumerate(segments[0:file_count]):
    file_name = "f{0}.wav".format(i)
    print "exporting ", file_name
    segment.export(file_name, format = 'wav') 
```

```python
audio_track = AudioSegment.from_mp3("drum.mp3")
segments = make_chunks(audio_track, segment_len)
```

```python
# Export all segments as .wave files

for i, segment in enumerate(segments[0:file_count]):
    file_name = "d{0}.wav".format(i)
    print "exporting ", file_name
    segment.export(file_name, format = 'wav') 
```
